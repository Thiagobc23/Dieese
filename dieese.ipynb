```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('dieese.csv')
df.head()
```

```python
from pandas.plotting import register_matplotlib_converters

df['DATE'] = df['MONTH'].map(str)+ '-' +df['YEAR'].map(str)
df['DATE'] = pd.to_datetime(df.DATE)

#df_temp = df[(df['YEAR']<2004) & (df['YEAR']>1999)]

fig, ax = plt.subplots(1,figsize=(20,10))
ax.set_facecolor('#1BBD10')
df_temp = df.sort_values(by='DATE').copy()
plt.plot(df_temp.DATE, df_temp.DIEESE,color='#F2CB3E')
plt.plot(df_temp.DATE, df_temp.MIN_WAGE, color='#105DBD')
ax.fill_between(df_temp.DATE, df_temp.DIEESE, df_temp.MIN_WAGE, color='#F2CB3E')
ax.fill_between(df_temp.DATE, df_temp.MIN_WAGE, 0, color='#105DBD')
ax.legend(['LIVING WAGE','MINIMUN WAGE'])
plt.grid(which='major', axis='y', alpha = 1, color='white')
ax.set_ylabel('MONTHLY VALUES IN BRL\n(220 HRS WORKED)', fontsize=18)
ax.set_xlabel('YEAR', fontsize=24)
ax.set_xticks(['1995-01-01','2000-01-01','2005-01-01','2010-01-01','2015-01-01','2019-01-01'])
plt.ylim((0, 4500))
plt.xlim(('1994-06-30', '2019-04-01'))
plt.title('LIVING WAGE VS MINIMUN WAGE IN BRAZIL', fontsize=22)
ax.tick_params(axis='both', which='major', labelsize=16)
fig.savefig('Brazil.png')
```
